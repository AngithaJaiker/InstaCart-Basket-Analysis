```python
# Create an outlier function to identify the outliers for our price
def find_outliers_IQR(ords_prods_merge):

   q1 = ords_prods_merge.quantile(0.25)

   q3 = ords_prods_merge.quantile(0.75)

   IQR = q3-q1

   outliers = ords_prods_merge[((ords_prods_merge<(q1-1.5*IQR)) | (ords_prods_merge>(q3+1.5*IQR)))]

   return outliers
```

```python
outliers = find_outliers_IQR(ords_prods_merge_subset['prices'])
```

```python
print('number of outliers: '+ str(len(outliers)))

print('max outlier value: '+ str(outliers.max()))

print('min outlier value: '+ str(outliers.min()))
```
